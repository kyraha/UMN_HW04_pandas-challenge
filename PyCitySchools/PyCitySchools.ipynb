```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# First cache some scalar counts
total_students = school_data_complete["Student ID"].count()

# And then cache different perspectives of the student data
passed_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
passed_read = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
passed_both = school_data_complete.loc[
    (school_data_complete['math_score'] >= 70) &
    (school_data_complete['reading_score'] >= 70)]
```

```python
# Compose a single row data frame with all the stats for the district
overall_stats = pd.DataFrame({
  "N of schols":   school_data["School ID"].count(),
  "N of students": total_students,
  "Total budget":  school_data['budget'].sum(),
  "Avg Math":      school_data_complete['math_score'].mean(),
  "Avg Reading":   school_data_complete['reading_score'].mean(),
  "Pass Math":     passed_math["Student ID"].count() / total_students,
  "Pass Reading":  passed_read["Student ID"].count() / total_students,
  "Pass All":      passed_both["Student ID"].count() / total_students
}, index=[0])

# Do some grooming with formats for pretty output
overall_stats['N of students'] = overall_stats['N of students'].map("{:,}".format)
overall_stats['Total budget']  = overall_stats['Total budget'].map("${:,}".format)
overall_stats['Avg Math']      = overall_stats['Avg Math'].map("{:.3f}".format)
overall_stats['Avg Reading']   = overall_stats['Avg Reading'].map("{:.3f}".format)
overall_stats['Pass Math']     = overall_stats['Pass Math'].map("{:.2%}".format)
overall_stats['Pass Reading']  = overall_stats['Pass Reading'].map("{:.2%}".format)
overall_stats['Pass All']      = overall_stats['Pass All'].map("{:.2%}".format)

# What do we got here..
overall_stats
```
